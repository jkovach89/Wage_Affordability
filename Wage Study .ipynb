```python
# Histogram - summarize number of states above 33 and 42 percent
wage_1968 = wage.loc[lambda yr: wage['Year']==1968]
wage_2016 = wage.loc[lambda yr: wage['Year']==2016]
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.hist(wage_1968['Perc.wage.33'], bins=50, color='c')
plt.axvline(0.333, color='b', linestyle='dashed', linewidth=2)
plt.axvline(0.419, color='b', linestyle='dashed', linewidth=2)
plt.title('Distribution of wages - 1968')

plt.subplot(1, 2, 2)
plt.hist(wage_2016['Perc.wage.33'], bins=50, color='c')
plt.axvline(0.333, color='b', linestyle='dashed', linewidth=2)
plt.axvline(0.419, color='b', linestyle='dashed', linewidth=2)
plt.title('Distribution of wages - 2017')

plt.show()
```
